```python
# First period: April 1854 - March 1855
months_54_55 = ['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar']
disease_54_55 = [1, 12, 11, 359, 828, 788, 503, 844, 1725, 2761, 2120, 1207]
wounds_54_55 = [0, 0, 0, 0, 1, 81, 132, 287, 114, 83, 42, 32]
other_54_55 = [5, 9, 6, 23, 30, 70, 128, 106, 131, 324, 361, 172]
```

```python
# Second period: April 1855 - March 1856
months_55_56 = ['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar']
disease_55_56 = [477, 508, 802, 382, 483, 189, 128, 178, 91, 42, 24, 15]
wounds_55_56 = [48, 32, 31, 30, 41, 47, 36, 19, 15, 2, 0, 0]
other_55_56 = [57, 37, 31, 33, 55, 37, 33, 18, 15, 3, 2, 4]
```

```python
# Create figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 9), 
                                subplot_kw={'projection': 'polar'})

# Colors
colors = {
    'disease': '#1f77b4',  # blue (center)
    'wounds': '#d62728',    # red (middle) 
    'other': '#2c2c2c'      # black (outer)
}

def create_coxcomb(ax, data_disease, data_wounds, data_other, months, title):
    """Create Nightingale coxcomb with original stacking: blue (center), red, black (outer)"""
    
    N = len(months)
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
    
    # Width of each wedge (30 degrees for monthly data)
    width = 2 * np.pi / N
    
    # Calculate radii - in Nightingale's plot, AREA is proportional to deaths
    # So radius = sqrt(deaths) * scale factor
    # We'll scale to make the plots visually comparable
    scale_factor = 15  # Adjust this to make the plot look good
    
    r_disease = np.sqrt(np.array(data_disease)) * scale_factor
    r_wounds = np.sqrt(np.array(data_wounds)) * scale_factor
    r_other = np.sqrt(np.array(data_other)) * scale_factor
    
    # Plot in original order: blue at center, then red, then black at outer edge
    # This means:
    # - Blue bars start at 0, height = r_disease
    # - Red bars start at r_disease, height = r_wounds
    # - Black bars start at r_disease + r_wounds, height = r_other
    
    # Plot disease (blue) - at center
    bars_disease = ax.bar(theta, r_disease, width=width, 
                          bottom=0, 
                          color=colors['disease'], 
                          edgecolor='white', linewidth=0.8,
                          label='Disease (Zymotic)',
                          alpha=0.9)
    
    # Plot wounds (red) - middle layer, starting where blue ends
    bars_wounds = ax.bar(theta, r_wounds, width=width, 
                         bottom=r_disease, 
                         color=colors['wounds'], 
                         edgecolor='white', linewidth=0.8,
                         label='Wounds & Injuries',
                         alpha=0.9)
    
    # Plot other (black) - outer layer, starting where blue + red ends
    bars_other = ax.bar(theta, r_other, width=width, 
                        bottom=r_disease + r_wounds, 
                        color=colors['other'], 
                        edgecolor='white', linewidth=0.8,
                        label='Other Causes',
                        alpha=0.9)
    
    # Customize the plot to match Nightingale's original orientation
    ax.set_theta_zero_location('N')  # North (top) is April (start)
    ax.set_theta_direction(-1)  # Clockwise
    
    # Set month labels - position them slightly outward
    ax.set_xticks(theta)
    ax.set_xticklabels(months, size=11, fontweight='bold')
    
    # Move the month labels outward so they don't overlap with the diagram
    ax.tick_params(axis='x', pad=15)
    
    # Remove radial labels and grid
    ax.set_yticklabels([])
    ax.set_rgrids([])
    ax.grid(False)
    
    # Remove the outer spine
    ax.spines['polar'].set_visible(False)
    
    # Add title
    ax.set_title(title, size=15, fontweight='bold', pad=25)
    
    return bars_disease, bars_wounds, bars_other

# Create first plot (1854-55)
create_coxcomb(ax1, disease_54_55, wounds_54_55, other_54_55, months_54_55,
               'APRIL 1854 TO MARCH 1855\n(Pre-Sanitation Reforms)')

# Create second plot (1855-56) 
create_coxcomb(ax2, disease_55_56, wounds_55_56, other_55_56, months_55_56,
               'APRIL 1855 TO MARCH 1856\n(Post-Sanitation Reforms)')

# Create a single legend
legend_elements = [
    mpatches.Patch(color=colors['disease'], label='Disease (Zymotic) - Blue', alpha=0.9),
    mpatches.Patch(color=colors['wounds'], label='Wounds & Injuries - Red', alpha=0.9),
    mpatches.Patch(color=colors['other'], label='Other Causes - Black', alpha=0.9)
]

fig.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, -0.02), 
           ncol=3, fontsize=12, frameon=False)

# Add main title - inspired by Nightingale's original
fig.suptitle("FLORENCE NIGHTINGALE'S COXCOMB PLOT\n" +
             "DIAGRAM OF THE CAUSES OF MORTALITY IN THE ARMY IN THE EAST",
             size=16, fontweight='bold', y=0.98)

# Add descriptive text like the original
description = ("The areas of the blue, red & black wedges are each measured from the centre as the common centre.\n"
               "The blue wedges measured from the centre represent the deaths from preventable or zymotic diseases,\n"
               "the red wedges measured from the centre the deaths from wounds, & the black wedges measured from\n"
               "the centre the deaths from all other causes during each month.")

fig.text(0.5, 0.01, description, ha='center', fontsize=10, 
         style='italic', wrap=True)

# Add data summary
total_deaths_pre = sum(disease_54_55) + sum(wounds_54_55) + sum(other_54_55)
total_deaths_post = sum(disease_55_56) + sum(wounds_55_56) + sum(other_55_56)
reduction = total_deaths_pre - total_deaths_post
percent_reduction = (reduction / total_deaths_pre) * 100

summary = f"Total Deaths (1854-55): {total_deaths_pre:,}   |   Total Deaths (1855-56): {total_deaths_post:,}   |   Reduction: {reduction:,} ({percent_reduction:.1f}%)"
fig.text(0.5, 0.94, summary, ha='center', fontsize=11, fontweight='bold', color='darkgreen')

plt.tight_layout()
plt.subplots_adjust(bottom=0.15, top=0.9)
plt.show()

# Print detailed statistics
print("\n" + "="*70)
print("NIGHTINGALE DATA SUMMARY")
print("="*70)

print("\nAPRIL 1854 - MARCH 1855 (PRE-SANITATION):")
print(f"   Disease deaths: {sum(disease_54_55):,} ({(sum(disease_54_55)/total_deaths_pre)*100:.1f}%)")
print(f"   Wound deaths:    {sum(wounds_54_55):,} ({(sum(wounds_54_55)/total_deaths_pre)*100:.1f}%)")
print(f"   Other deaths:    {sum(other_54_55):,} ({(sum(other_54_55)/total_deaths_pre)*100:.1f}%)")
print(f"   TOTAL:           {total_deaths_pre:,}")

print("\nAPRIL 1855 - MARCH 1856 (POST-SANITATION):")
print(f"   Disease deaths: {sum(disease_55_56):,} ({(sum(disease_55_56)/total_deaths_post)*100:.1f}%)")
print(f"   Wound deaths:    {sum(wounds_55_56):,} ({(sum(wounds_55_56)/total_deaths_post)*100:.1f}%)")
print(f"   Other deaths:    {sum(other_55_56):,} ({(sum(other_55_56)/total_deaths_post)*100:.1f}%)")
print(f"   TOTAL:           {total_deaths_post:,}")

print("\nTHE IMPACT OF SANITATION REFORMS:")
print(f"   Deaths reduced by {reduction:,} ({percent_reduction:.1f}%)")
print(f"   Disease deaths alone reduced by {sum(disease_54_55)-sum(disease_55_56):,} deaths")
```
